# Notebook 05: Final Load Prep (Tableau-Ready Export)
**Objective:** Compute all KPIs and export aggregated tables for Tableau dashboard.

```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('../data/processed/ecommerce_cleaned.csv')
os.makedirs('../data/processed', exist_ok=True)
print('Loaded:', df.shape)
```

```python
# ── KPI Summary Table ────────────────────────────────────
kpi_df = pd.DataFrame([
    {'KPI': 'Total Revenue', 'Value': df['total_amount'].sum(), 'Unit': 'INR'},
    {'KPI': 'Total Orders', 'Value': len(df), 'Unit': 'count'},
    {'KPI': 'Avg Order Value', 'Value': df['total_amount'].mean().round(2), 'Unit': 'INR'},
    {'KPI': 'Avg Customer Rating', 'Value': df['rating'].mean().round(3), 'Unit': 'score'},
    {'KPI': 'Avg Delivery Days', 'Value': df['delivery_days'].mean().round(2), 'Unit': 'days'},
    {'KPI': 'High-Value Orders', 'Value': df['is_high_value'].sum(), 'Unit': 'count'},
    {'KPI': 'Fast Delivery Orders', 'Value': df['fast_delivery'].sum(), 'Unit': 'count'},
    {'KPI': 'High-Rating Orders', 'Value': df['high_rating'].sum(), 'Unit': 'count'},
])
kpi_df.to_csv('../data/processed/kpi_summary.csv', index=False)
print(kpi_df.to_string(index=False))
```

```python
# ── Revenue by Category ──────────────────────────────────
rev_cat = df.groupby('product_category').agg(
    total_revenue=('total_amount', 'sum'),
    order_count=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean'),
    avg_rating=('rating', 'mean'),
    avg_delivery_days=('delivery_days', 'mean')
).round(2).reset_index()
rev_cat.to_csv('../data/processed/revenue_by_category.csv', index=False)
print(rev_cat)
```

```python
# ── Revenue by City ──────────────────────────────────────
rev_city = df.groupby('city').agg(
    total_revenue=('total_amount', 'sum'),
    order_count=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean'),
    avg_rating=('rating', 'mean')
).round(2).reset_index()
rev_city.to_csv('../data/processed/revenue_by_city.csv', index=False)
print(rev_city)
```

```python
# ── Monthly Trend ────────────────────────────────────────
monthly = df.groupby(['month', 'month_name']).agg(
    total_revenue=('total_amount', 'sum'),
    order_count=('order_id', 'count'),
    avg_rating=('rating', 'mean')
).round(2).reset_index().sort_values('month')
monthly.to_csv('../data/processed/monthly_trend.csv', index=False)
print(monthly)
```

```python
# ── Category x City Revenue Matrix ──────────────────────
matrix = df.pivot_table(
    values='total_amount', index='city', columns='product_category', aggfunc='sum'
).round(0)
matrix.to_csv('../data/processed/category_city_matrix.csv')
print(matrix)
```
