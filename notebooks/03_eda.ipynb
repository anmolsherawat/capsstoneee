# Notebook 03: Exploratory Data Analysis
**Objective:** Understand distributions, trends, and patterns across categories, cities, and time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

df = pd.read_csv('../data/processed/ecommerce_cleaned.csv')
print(f'Loaded: {df.shape}')
```

```python
# ── KPI Summary ──────────────────────────────────────────
kpis = {
    'Total Revenue (₹)': df['total_amount'].sum(),
    'Avg Order Value (₹)': df['total_amount'].mean().round(0),
    'Avg Customer Rating': df['rating'].mean().round(2),
    'Avg Delivery Days': df['delivery_days'].mean().round(2),
    'Total Orders': len(df),
    'High-Value Orders (top 10%)': df['is_high_value'].sum()
}
for k, v in kpis.items():
    print(f'{k}: {v:,.2f}' if isinstance(v, float) else f'{k}: {v:,}')
```

```python
# ── Revenue by Category ──────────────────────────────────
rev_cat = df.groupby('product_category')['total_amount'].sum().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(8, 4))
rev_cat.plot(kind='bar', color=['#1C7293', '#028090', '#02C39A', '#0D9488'], ax=ax)
ax.set_title('Revenue by Product Category', fontsize=14, fontweight='bold')
ax.set_xlabel('')
ax.set_ylabel('Total Revenue (₹)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'₹{x/1e7:.1f}Cr'))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
print(rev_cat.to_dict())
```

```python
# ── Revenue by City ──────────────────────────────────────
rev_city = df.groupby('city')['total_amount'].sum().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(8, 4))
rev_city.plot(kind='barh', color='#065A82', ax=ax)
ax.set_title('Revenue by City', fontsize=14, fontweight='bold')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'₹{x/1e7:.1f}Cr'))
plt.tight_layout()
plt.show()
```

```python
# ── Monthly Revenue Trend ────────────────────────────────
monthly = df.groupby('month_name')['total_amount'].sum().reindex(['January', 'February', 'March'])
fig, ax = plt.subplots(figsize=(8, 4))
monthly.plot(kind='line', marker='o', color='#028090', linewidth=2.5, ax=ax)
ax.set_title('Monthly Revenue Trend (Jan–Mar 2024)', fontsize=14, fontweight='bold')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'₹{x/1e7:.1f}Cr'))
plt.tight_layout()
plt.show()
```

```python
# ── Rating Distribution ──────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['rating'].value_counts().sort_index().plot(kind='bar', color='#1C7293', ax=axes[0])
axes[0].set_title('Overall Rating Distribution')
axes[0].set_xlabel('Rating')

df.groupby('product_category')['rating'].mean().plot(kind='bar', color='#02C39A', ax=axes[1])
axes[1].set_title('Avg Rating by Category')
axes[1].set_xlabel('')
axes[1].set_ylim(0, 5)
plt.tight_layout()
plt.show()
```

```python
# ── Delivery Days Distribution ──────────────────────────
fig, ax = plt.subplots(figsize=(8, 4))
df['delivery_days'].value_counts().sort_index().plot(kind='bar', color='#065A82', ax=ax)
ax.set_title('Delivery Days Distribution')
ax.set_xlabel('Days')
plt.tight_layout()
plt.show()
print('Fast delivery (<= 3 days):', df['fast_delivery'].mean()*100, '%')
```

```python
# ── Payment Method Split ─────────────────────────────────
pay = df[df['payment_method'] != 'Unknown']['payment_method'].value_counts()
pay.plot(kind='pie', autopct='%1.1f%%', colors=['#1C7293','#028090','#02C39A'])
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
